```python
import os
import cv2
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
from imblearn.over_sampling import SMOTE
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten, Dense
import matplotlib.pyplot as plt
```

```python
# Set your dataset path
dataset_path = "combined_data"
```

```python
# Function to load and preprocess images
def load_images_and_labels(dataset_path):
    paths, labels = [], []

    for emotion_folder in os.listdir(dataset_path):
        emotion_path = os.path.join(dataset_path, emotion_folder)
        if os.path.isdir(emotion_path):
            for file in os.listdir(emotion_path):
                file_path = os.path.join(emotion_path, file)
                if os.path.isfile(file_path):
                    paths.append(file_path)
                    labels.append(emotion_folder)

    df = pd.DataFrame({'path': paths, 'label': labels})
    df['img'] = df['path'].apply(lambda path: cv2.imread(path, cv2.IMREAD_GRAYSCALE))  # Convert to grayscale

    return df
```

```python
# Load and preprocess images
df = load_images_and_labels(dataset_path)
```

```python
# Shuffle the dataframe
df = shuffle(df, random_state=42)

# Split the data into features (X) and labels (y)
X = np.stack(df['img'])
y = df['label']
```

```python
# Encode labels using label_encoder
label_encoder = preprocessing.LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
```

```python
# Reshape X for CNN
X = X / 255.0  # Normalize pixel values
X = X.reshape((X.shape[0], 48, 48, 1))  # Keep 1 channel for grayscale
```

```python
# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)
```

```python
# Apply SMOTE for class imbalance
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train.reshape((X_train.shape[0], -1)), y_train)
```

```python
# Reshape X_train_resampled back to its original shape
X_train_resampled = X_train_resampled.reshape((X_train_resampled.shape[0], 48, 48, 1))
```

```python
# Define the CNN model
model = Sequential()

model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(48, 48, 1), name='conv1'))
model.add(Conv2D(64, kernel_size=(3, 3), activation='relu', name='conv2'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(128, kernel_size=(3, 3), activation='relu', name='conv3'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(256, kernel_size=(3, 3), activation='relu', name='conv4'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(7, activation='softmax'))
```

```python
# Compile the model
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
history = model.fit(X_train_resampled, y_train_resampled, epochs=2, validation_data=(X_test, y_test))
```

```python
from keras.callbacks import History

# Initialize History callback
history = History()

# Train the model with the History callback
model.fit(X_train_resampled, y_train_resampled, epochs=1, validation_data=(X_test, y_test), callbacks=[history])

# Plot training history
plt.figure(figsize=(12, 4))

# Plot training accuracy
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Plot training loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```
